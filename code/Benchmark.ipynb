```python
import pandas as pd
import numpy as np

import Benchmark as bm
```

```python
# Load datasets

real = pd.read_csv('final.csv', usecols= ['anchor_age','Heart Rate', 'O2 saturation pulseoxymetry', 'Respiratory Rate'])

# Drop outliers in the O2 Saturation pulseoxymetry column
real = real[(real['O2 saturation pulseoxymetry'] != 811100.0) & (real['O2 saturation pulseoxymetry'] != 9795.0)]

synthetic = pd.read_csv('synthetic data.csv', usecols =['anchor_age','Heart Rate', 'O2 saturation pulseoxymetry', 'Respiratory Rate'])
```

```python
OG_Corr, Synth_Corr, Corr_Diff = bm.correlation_heatmaps(real, synthetic)

corr_diff_sum = bm.Corr_Diff_Sum(Corr_Diff)

print("Sum of the correlation differences: ", corr_diff_sum)
```
